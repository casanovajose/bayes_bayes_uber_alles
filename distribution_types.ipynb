```python
%matplotlib inline

import json
import matplotlib
s = json.load(open("styles/bmh_matplotlibrc.json"))

s["axes.prop_cycle"] = matplotlib.cycler(color=s["axes.prop_cycle"])
matplotlib.rcParams.update(s)

from IPython.core.pylabtools import figsize
import numpy as np
import matplotlib.pyplot as plt
figsize(11, 9)

import scipy.stats as stats
```

Discrete - Poisson

```python
a = np.arange(16)
poi = stats.poisson
lambda_ = [1.5, 4.25, 12]
colours = ["#348ABD", "#A60628", "#666666"]

plt.bar(a, poi.pmf(a, lambda_[0]), color=colours[0], alpha=0.5,
        label="$\lambda = %.1f$" % lambda_[0])

plt.bar(a, poi.pmf(a, lambda_[1]), color=colours[1], alpha=0.5,
        label="$\lambda = %.1f$" % lambda_[1])

plt.bar(a, poi.pmf(a, lambda_[2]), color=colours[2], alpha=0.5,
        label="$\lambda = %.1f$" % lambda_[2])

plt.xticks(a + 0.4, a)
plt.legend()
plt.ylabel("probability of $k$")
plt.xlabel("$k$")
plt.title("PMF of a Poisson random variable; differeing $\lambda$ values")
```

```python
Continuos - Exponential
```

```python
a = np.linspace(0, 4, 100)
expo = stats.expon
lambda_ = [0.5, 1, 2]

for l, c in zip(lambda_, colours):
    plt.plot(a, expo.pdf(a, scale=1./l), lw=4,
             color=c, label="$\lambda = %.1f$" % l)
    

plt.legend()
plt.ylabel("PDF at $z$")
plt.xlabel("$z$")
plt.ylim(0, 2)
plt.title("PDF of an Exponential random variable; differeing $\lambda$ values")
```
